```python
import cv2
import mediapipe as mp
import numpy as np
import os

# 1. AYARLAR: Hangi kelimeleri çekeceğiz?
DATA_PATH = os.path.join('My_VitalSign_Data') 
# Başlangıç için sadece 3 kelimeyle küçük bir deneme yapalım:
actions = np.array(['help', 'doctor', 'emergency']) 
no_sequences = 10 # Her kelime için 10'ar video çekeceğiz
sequence_length = 30 # Her video 30 kare sürecek

# Klasörleri otomatik oluştur
for action in actions: 
    for sequence in range(no_sequences):
        try: 
            os.makedirs(os.path.join(DATA_PATH, action, str(sequence)))
        except:
            pass

# 2. MEDIAPIPE KURULUMU
mp_holistic = mp.solutions.holistic 
mp_drawing = mp.solutions.drawing_utils 

def mediapipe_detection(image, model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False                  
    results = model.process(image)                 
    image.flags.writeable = True                   
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) 
    return image, results

def extract_landmarks(results):
    # Sol El
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    # Sağ El
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    
    # Sadece X ve Y koordinatlarını alarak boyutu modelimize göre (108) ayarlıyoruz
    lh_xy = lh.reshape(-1, 3)[:, :2].flatten() # 21 nokta * 2 = 42
    rh_xy = rh.reshape(-1, 3)[:, :2].flatten() # 21 nokta * 2 = 42
    
    # Yüzden de en belirgin 12 noktayı (göz, ağız çevresi) alıyoruz
    face = np.array([[res.x, res.y] for res in results.face_landmarks.landmark[:12]]).flatten() if results.face_landmarks else np.zeros(12*2)
    
    # Toplam: 42 + 42 + 24 = TAM 108 BOYUT!
    return np.concatenate([lh_xy, rh_xy, face])

# 3. KAYIT DÖNGÜSÜ
print("Starting camera... Please get ready!")
cap = cv2.VideoCapture(0)
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    for action in actions:
        for sequence in range(no_sequences):
            for frame_num in range(sequence_length):
                ret, frame = cap.read()
                image, results = mediapipe_detection(frame, holistic)
                
                # EKRAN ÇİZİMLERİ (Yüz çizimi eklendi!)
                if results.face_landmarks:
                    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS, 
                                             mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1),
                                             mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1))
                if results.left_hand_landmarks:
                    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
                if results.right_hand_landmarks:
                    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

                # Ekrana bilgilendirme yazıları yazalım
                if frame_num == 0: 
                    cv2.putText(image, 'GET READY...', (120,200), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0,255, 0), 4, cv2.LINE_AA)
                    cv2.putText(image, f'Recording {action.upper()} | Video: {sequence+1}/{no_sequences}', (15,30), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
                    cv2.imshow('VitalSign Data Collector', image)
                    cv2.waitKey(4000) # Hazırlanma süresi 4 saniyeye çıkarıldı
                else: 
                    cv2.putText(image, f'Recording {action.upper()} | Video: {sequence+1}/{no_sequences}', (15,30), 
                               cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
                    cv2.imshow('VitalSign Data Collector', image)

                # Landmarkları çıkar ve kaydet
                keypoints = extract_landmarks(results)
                npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_num))
                np.save(npy_path, keypoints)

                # Kapatmak istersen klavyeden 'q' tuşuna bas
                if cv2.waitKey(10) & 0xFF == ord('q'):
                    break
                    
    cap.release()
    cv2.destroyAllWindows()
    print("Recording completed successfully!")
```
